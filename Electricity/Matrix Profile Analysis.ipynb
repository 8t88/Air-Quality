```python
import pandas as pd
electricity_pd_df = pd.read_pickle("cal_electricity_mwh.pickle")
electricity_pd_df['date_hr'] = pd.to_datetime(electricity_pd_df['date_hr']) #.dt.strftime('%Y-%m-%d-%H')
electricity_pd_df['megawatthours'] = electricity_pd_df['megawatthours'].astype(float)
sce_electricity_pd_df = electricity_pd_df[electricity_pd_df['subregion_abbrv']=='SCE'].sort_values('date_hr')
```

```python
X=sce_electricity_pd_df['date_hr']
ts=sce_electricity_pd_df['megawatthours']

plt.figure(figsize=(20,5))
plt.plot(X, ts)
plt.title('Electricity Time Series')
plt.show()
```
